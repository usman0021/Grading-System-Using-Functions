### Function to calculate grade based on marks

```python
def calculate_grade(marks):
    if marks >= 90:
        return "A+"
    elif marks >= 80:
        return "A"
    elif marks >= 70:
        return "B"
    elif marks >= 60:
        return "C"
    elif marks >= 50:
        return "D"
    else:
        return "F"
```

```python
def display_result(name, marks):
    grade = calculate_grade(marks)
    print("Student Result")
    print("Student Name:", name)
    print("Marks Obtained:", marks)
    print("Grade:", grade)
```

```python
student_name = input("Enter student name: ")
student_marks = int(input("Enter student marks (0-100): "))
display_result(student_name, student_marks)
```

```python
###### Data Base
users = {
    "uxman00021@gmail.com": "6543210",
    "admin@gmail.com": "12345"
}

add_user = int(input("How many new users do you want to add? "))

for i in range(add_user):
    email = input(f"Enter new user email #{i+1}: ")
    password = input(f"Enter password for {email}: ")
    users[email] = password

print("Updated user database:")
print(users)

# GRADING SYSTEM FUNCTIONS

def calculate_grade(marks):
    """Return grade based on marks."""
    if marks >= 90:
        return "A+"
    elif marks >= 80:
        return "A"
    elif marks >= 70:
        return "B"
    elif marks >= 60:
        return "C"
    elif marks >= 50:
        return "D"
    else:
        return "F"


def display_grade(name, marks):
    """Print student details + grade."""
    grade = calculate_grade(marks)
    print("Student Result")
    print("Student Name:", name)
    print("Marks Obtained:", marks)
    print("Grade:", grade)


# LOGIN SYSTEM (WHILE LOOP)

print("\nLogin System (WHILE LOOP)")
login_success = False

while not login_success:
    Email = input("Enter your email: ")

    if '@' in Email:
        Password = input("Enter your password: ")

        if Email in users and users[Email] == Password:
            print("Login Successful! Welcome,", Email)
            login_success = True
        else:
            print("Incorrect information. Please try again.\n")
    else:
        print("Wrong email format")


# MENU AFTER LOGIN (LOGOUT SYSTEM)

while True:
    print("MAIN MENU")
    print("1. Add Student Grade")
    print("2. Logout")

    choice = input("Enter your choice (1 or 2): ")

    if choice == "1":
        student_name = input("Enter student name: ")
        student_marks = int(input("Enter student marks (0-100): "))
        display_grade(student_name, student_marks)

    elif choice == "2":
        print("logged out succesful")
        break

    else:
        print("Invalid choice! Please enter 1 or 2.")
```
